## Data Load

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset/autos_imports-85.csv')
# df.columns = headers
df.head()
```

## Data Preprocessing
### What is Missing Value

- Occured when no data value is stored for a variable (feature) in an observation.
- Missing value is represented by '?', 'N/A', 0, or just a blank cell.

### Handling Missing Value 
- Drop the missing values
  - Drop the variable
  - Drop the data entry (preferable)
- Replace the missing value (Recommended)
  - Replace it with an average (of similar datapoints)
  - Replace it by frequency (mode)
  - Replace it based on other functions
- Leave it as missing data

```python
df.head()
```

#### Replace Missing Value Identifier 

Notice that we have "?" as the missing value identifier. We need to replace it with "None" so that we can identify missing values using `.isna()`

```python
df.replace('?', np.nan, inplace=True)
df.head()
```

```python
columns = df.columns.to_list()
for col in columns:
  print(df[col].isna().value_counts())
```

#### Drop the Missing Values

Now, we have 4 missing values in the price columns. Let's drop it!

```python
# Drop the missing value
df_master = df.dropna(subset= ['price'], axis=0)
df_master.head()
```

#### Replace the Missing Values with Mean

```python
df.columns.to_list()
```

```python
# Let's change the data type for numerical columns
num_columns = ['normalized-losses', ' wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size',  'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
# num_columns = ['normalized-losses']
for col in num_columns:
  df_master[col] = df_master[col].astype(np.float64)
# df_master['normalized-losses'] = df_master['normalized-losses'].astype('Int64')
df_master.info()
```

```python
df_master.head()
```

```python
mean = df_master['normalized-losses'].mean()
df_master['normalized-losses'].fillna(mean, inplace=True)
df_master.head()
```
